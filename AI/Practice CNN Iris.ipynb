```python
import pandas as pd
import numpy as np
from tensorflow import keras
from tensorflow import one_hot
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
print('Imports are working')
```

```python
if __name__ == "__main__":
    
    
    
    # Person 1_________________________________GET DATA________________________________________________________
    
    
    # Here is where you will train the data from the Iris.csv files.

    dataset = pd.read_csv("Iris.csv")
    
        
    
    # Here is where you will get the values for X and Y.
    
    y = np.array(dataset["Species"]) #Gets the integer values
    X = np.array(dataset[["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]])

    
    
    
    
    
    
    
    
    # Person 2_________________________________SPLIT DATA_______________________________________________________

    
    
    # Here is where you will SPLIT up the train and test data.
    
    # Hint: One of the blanks is = train_test_split
    
    # Hint: What is a good test size as a decimal [0,1], smaller or higher? 
        # what will it do to the train size?

    x_train, x_test, y_train, y_test =train_test_split(X, y, test_size = 0.3, random_state=42)
    
    
    
    #One hot encodes, better for training. 
    y_train = one_hot(y_train,3)
    y_test = one_hot(y_test,3)
    
    
    
    
    
    

    # Person 3_________________________________CREATE MODEL_____________________________________________________

    # Hints: Answer Key
        # relu          (Additional hint, this is used more than once!)
        # softmax
        # add           (Additional hint, this is used more than once!)
        # Dense
        # Sequential()

    #Creates Sequential Model
    model = keras.models.Sequential()
    model.add(keras.layers.Dense(100,activation = "relu"))
    model.add(keras.layers.Dense(10,activation = "relu"))
    model.add(keras.layers.Dense(3,activation = "softmax"))

                     
                     
                     
                     
    # Person 4_________________________________COMPILE MODEL_____________________________________________________

    # Hints: Answer Key
                     
        # ["accuracy"]
        # "categorical_crossentropy"
        # epochs
        # x_train
        # y_train
        # x_test
        # y_test
        
        # 'Samsung Sam approves this model!'
                     
                     
                     
                     
    model.compile(loss = "categorical_crossentropy",optimizer = "adam" , metrics = ["accuracy"])
    model.fit(x_train, y_train, epochs = 50 , batch_size = 12) 

    model.summary()

    model.evaluate(x_test,y_test)  #Tests the model
         
    print('Samsung Sam approves this model!!!')
```
